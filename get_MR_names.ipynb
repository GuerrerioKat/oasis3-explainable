```python
import pandas as pd
import numpy as np

input_file = "OASIS3_MR_json.csv"
output_file = "oasis3-1per.csv"

df = pd.read_csv(input_file)
df
```

```python
random_seed = 42
df_one = (df.groupby("subject_id", group_keys=False).apply(lambda x: x.sample(1, random_state=random_seed)))
df_one['label']
```

```python
df_label = pd.DataFrame(df_one['label'])
df_label
```

```python
df_label.to_csv(output_file, index=False)

print(f"Saved {len(df_one)} rows to {output_file}")
```
